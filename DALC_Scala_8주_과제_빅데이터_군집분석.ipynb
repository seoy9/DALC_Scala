## 8-1 빅데이터 비지도 학습의 개념

### 1. 비지도 학습
</br>

[ __머신러닝__ ]
+ 지도학습 : 레이블 데이터가 있는 것
+ 비지도 학습 : 레이블 데이터가 없는 것  
</br></br>

##### 1) 개요
[ __비지도학습__ ]
+ 머신러닝의 일종
+ 데이터 어떻게 구성되었는지를 알아내는 문제의 범주에 속함
+ 지도 학습과는 다르게 레이블이 없거나 참조하지 않음


 → __데이터 세트가 스스로 데이터를 설명하게 하는 것__
</br></br>

 [ __데이터 특징과 레이블__ ]
 + 대표적 비지도 학습 : 군집화, 자원 축소
 + 군집화 분석 : 특징 데이터를 문석하여 유사한 데이터들을 그룹으로 묶는, 군집화 하는 알고리즘 ( __구체적인 특성을 공유하는 군집을 찾는 것__ )
 + 차원 축소 : __데이터세트의 특징행렬을 좀 더 간결하게 표현하는 방법__

특징 : 기온, 습도 / 레이블 : 빈칸

기온      | 습도     | .
----------|----------|----------
20        | 60       | .
30        | 80       | .
30        | 70       | .
35        | 75       | .

+ 군집화 : 기온과 습도 두 개의 차원으로 구성된 특징 데이터를 유사한 것들끼리 묶는 방법
+ 차원 축소 : 기온과 습도 2개의 차원으로 표현되는 데이터를 하나의 특징행렬로 표현할 수 있다면 → 차원을 축소함
+ 특징행렬의 차원이 크지 않으므로 차원 축소의 효용이 높지 않지만, 특징행렬의 차원이 큰 경우에는 차원 축소의 효용이 상당히 높다고 할 수 있음 ( __특징행렬의 차원이 클 수록 차원 축소의 효용이 높음__ )
</br></br>

## 8-2 심화: K-평균 군집 분석

### 1. K-평균 알고리즘
</br>

##### 1) 개요
[ __최적의 군집화__ ]
+ 군집 중앙은 해당 군집에 속하는 모든 점의 산술 평균
+ 각 점은 다른 군집의 중앙보다 자신이 속한 군집의 중앙에 더 가까움
</br></br>

[ __표준 패키지 불러오기__ ]

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

[ __데이터 생성__ ]

```python
from sklearn.datasets.samples_generator import make_blobs
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);
```

##### 2) 절차
[ __모델 클래스 불러오기__ ]

```python
from sklearn.cluster import KMeans
```

[ __모델 인스턴스 생성 및 초모수 설정__ ]

```python
kmeans = KMeans(n_clusters=4) # 중심점의 개수를 설정하는 초모수
```

[ __모델 적합__ ]

```python
kmeans.fit(X)
```

[ __모델 적용__ ]

```python
y_kmeans = kmeans.predict(X)
```

[ __군집 중심 시각화__ ]

```python
centers = kmeans.cluster_centers_ # 중심점의 좌표는 모델의 cluster_centers_ 멤버 변수에 저장
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

[ __데이터 시각화__ ]

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis');
```

##### 3) 원리
[ __기댓값-최대화(E-M)__ ]
+ 관측되지 않는 잠재 변수에 의존하는 확률 모델에서 최대가능도를 갖는 모수의 추정값을 찾는 반복적인 알고리즘

1. 군집 중심을 임의로 추측한다.
2. 군집 중심의 위치가 수렴할 때까지 다음을 반복한다.  
- 기댓값 단계 : 각 점을 가장 가까운 군집 중심에 할당한다.  
- 최대화 단계 : 할당된 데이터들의 산술평균을 통해 새로운 군집 중심을 찾는다.
3. 수렴된 군집의 중심을 기준으로 군집을 할당한다.
![기댓값-최대화](https://github.com/seoy9/DALC_Scala/blob/7b2a27b6fd38710df38846ba61572340f7c76a3d/%EA%B8%B0%EB%8C%93%EA%B0%92%20%EC%B5%9C%EB%8C%80%ED%99%94.JPG?raw=true)
+ 1번: 1
+ 2번-기댓값 단계 : 2-1, 2-3, 2-5
+ 2번-최대화 단계 : 2-2, 2-4, 2-6
+ 3번 : 3
</br></br>

## 8-3 심화: 가우스 혼합 모델

### 1. 가우스 혼합 모델
</br>

##### 1) 등장 배경
[ __K-평균 알고리즘의 약점__ ]
+ K-평균 알고리즘의 장점
  + 사용이 편리
  + 직관적인 개념
+ K-평균 알고리즘의 단점
  + 유연성 부족 : 군집 중심 원형 경계를 사용한 군집 할당
  + 확정적 군집 할당 : 비확률적인 군집 할당 기준
</br></br>

##### 2) 정상적인 적용 결과 Vs. 비정상적인 적용 결과
[ __표준 패키지 불러오기__ ]

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

[ __정상적인 K-평균 군집 분석__ ]

```python
from sklearn.datasets.samples_generator import make_blobs
X, y_true = make_blobs(n_samples=400, centers=4, cluster_std=0.60, random_state=0)
# 데이터 샘플의 수 400개, 중심점의 개수 4개
X = X[:, ::-1]
plt.scatter(X[:, 0], X[:, 1], c='gray', alpha=0.4);
# 데이터의 시인성을 높이기 위해 두 번째 차원의 데이터 순서를 뒤집어서 시각화
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, random_state=0) # 군집 중심을 4개로 설정
clusters = kmeans.fit(X).predict(X)
plt.scatter(X[:, 0], X[:, 1], c=clusters, s=40, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red');
# 군집별로 다른 색상을 할당하여 구분, 중심점은 빨간색으로 표시
```
