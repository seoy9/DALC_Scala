# 중복 데이터 처리

* 다음 사진과 같이 데이터프레임을 생성해라

캡처.PNG

```python
df = pd.DataFrame({'A':['a', 'a', 'b', 'a', 'b'], 'B':[1, 1, 1, 2, 1], 'C':[1, 1, 2, 2, 2]})
print(df)
```

* 중복 데이터 확인해라(이전행과 현재행 비교)

```python
df_dup = df.duplicated()
print(df_dup)
```

* 중복 데이터를 삭제하고 고유한 값만을 가지 행들만 남겨라(원본에 저장)

```python
df2 = df.drop_duplicates()
print(df2)
```

* B열과 C열기준으로 중복 여부를 판단해라

```python
df3 = df.drop_duplicates(subset=['B', 'C'])
print(df3)
```
